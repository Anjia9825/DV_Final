```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
#!pip install yfinance
import yfinance as yf
import pandas as pd
```

```python
#Collection and Preprocessing of Data
BTC = yf.Ticker('BTC-USD')
ETH = yf.Ticker('ETH-USD')

BTC_hist = BTC.history(period="max", auto_adjust=True)
BTC_df = pd.DataFrame()
BTC_df['Date'] = BTC_hist.index
BTC_df['Close'] = BTC_hist['Close'].values
BTC_df['Open'] = BTC_hist['Open'].values
BTC_df['High'] = BTC_hist['High'].values
BTC_df['Low'] = BTC_hist['Low'].values
BTC_df['Volume'] = BTC_hist['Volume'].values
BTC_df=BTC_df.tail(1000)
BTC_df 
```

```python
BTC_df.to_csv('BTC_USD.csv', index=False)
```

```python
ETH_hist = ETH.history(period="max", auto_adjust=True)
ETH_df = pd.DataFrame()
ETH_df['Date'] = ETH_hist.index
ETH_df['Close'] = ETH_hist['Close'].values
ETH_df['Open'] = ETH_hist['Open'].values
ETH_df['High'] = ETH_hist['High'].values
ETH_df['Low'] = ETH_hist['Low'].values
ETH_df['Volume'] = ETH_hist['Volume'].values
ETH_df=ETH_df.tail(1000)
ETH_df 
```

```python
ETH_df.to_csv('ETH_USD.csv', index=False)
```

```python
Crypto_data = pd.concat([BTC_df.set_index('Date'), ETH_df.set_index('Date')],
                         axis=1, keys=['BTC', 'ETH'])

Crypto_data.columns = Crypto_data.columns.map('_'.join)
Crypto_data = Crypto_data.reset_index()
Crypto_data
```

```python
Crypto_data['average'] = (Crypto_data['BTC_Close'] + Crypto_data['ETH_Close']) / 2
#Crypto_data=Crypto_data.tail(1000)
Crypto_data
```

```python
Crypto_data['total_Volume'] = (Crypto_data['BTC_Volume'] +
                                Crypto_data['ETH_Volume'])

Crypto_data
```

```python
for crypto in ['BTC', 'ETH']:
    Crypto_data[f'{crypto}_percentage'] = (Crypto_data[f'{crypto}_Volume'] /
                                           Crypto_data['total_Volume']) * 100
```

```python
# Convert the data to CSV format
Crypto_data.to_csv('Crypto_data.csv', index=False)
```
